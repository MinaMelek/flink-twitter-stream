### Apache Flink:
   Apache Flink is an open source engine for distributed stream and batch processing. In contrast with its previous counterpart (Apache Spark), Flink is a native stream processing engine engine and the batch processing is built on top of the streaming engine.
 <img src="https://ci.apache.org/projects/flink/flink-docs-release-1.7/page/img/navbar-brand-logo.jpg">
 
### Anatomy of a Flink Program:
   The Flink program is a DAG of operators (e.g., filtering, updating state, defining windows, aggregating) that transform  collections of data. The usual Flink program consists of:
   - Get an execution environment,
   - Get data from a source, 
   - Conduct your transformations on this data,
   - Send the result to a sink,
   - Trigger the program execution
### Flink basics:
#### Flink data sources:
   Sources are where Flink reads its input from. Flink provides several predefined sources such as file, socket and collection. Moreover, the user can implement his own custom sources such as a Kafka consumer. 
#### Stream transformations:
Flink introduces multiple transformations that can be combined into sophisticated dataflow topologies. These transformations (operators) include Map, FlatMap, Filter, KeyBy, Reduce, Window and many more. We refer the reader to <a href=https://ci.apache.org/projects/flink/flink-docs-release-1.7/dev/stream/operators/index.html> here</a> for more information.
#### Flink data sinks:
Sinks are where the data stream is consumed. Flink provides several predefined sinks such as files and sockets. Moreover, a user can define his own custom sinks as well.

The implementation of ```Splitter``` class is as the following:
```java 
public static class Splitter implements FlatMapFunction<String, Tuple2<String, Integer>> {
    @Override
    public void flatMap(String sentence, Collector<Tuple2<String, Integer>> out) throws Exception {
        for (String word: sentence.split(" ")) {
            out.collect(new Tuple2<String, Integer>(word, 1));
        }
    }
}
```

Print out the results and invoke the execution.

``` java
dataStream.print();
env.execute("Window WordCount");
```
